### –û–ø–µ—Ä–∞—Ü–∏–∏ –Ω–∞–¥ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏

- –û—Å–Ω–æ–≤–Ω—ã–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã:
    - `+`
    - `-`
    - `/`
    - `*`
- –û–ø–µ—Ä–∞—Ç–æ—Ä—ã —Å—Ä–∞–≤–Ω–µ–Ω–∏—è (–∫–∞–∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç —Ç–∏–ø `bool`):
    -  `>` - –±–æ–ª—å—à–µ, `>=` - –±–æ–ª—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ
    -  `<` - –º–µ–Ω—å—à–µ, `<=` - –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ
    -  `==` - –ø—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ (–≤–æ–∑—Ä–≤–∞—â–∞–µ—Ç –¥–≤–∞ –µ—Å–ª–∏ –∑–Ω–∞—á–µ–Ω–∏–µ –æ–¥–∏–Ω–∞–∫–æ–≤–æ)
    -  `!=` - –Ω–µ —Ä–∞–≤–Ω–æ
- –õ–æ–≥–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã:
    - `and`
    - `or`
    - `not`
- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã**:
    - `x**n` - `x` –≤ —Å—Ç–µ–ø–µ–Ω–∏ `n`
    - `//` - –¥–µ–ª–µ–Ω–∏–µ –Ω–∞—Ü–µ–ª–æ
    - `%` - –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è

````python
print('10 + 2 =', 10 + 2)
print('10 - 2 =', 10 - 2)
print('10/2 =', 10/2) # –æ–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ —á—Ç–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç —É–∂–µ –Ω–µ integer
print('10*2 =', 10*2)
print('10 > 2 =', 10 > 2)
print('10 < 2 =', 10 < 2)
print('10 == 2 =', 10 == 2)
print('10 != 2 =', 10 != 2)
print('True and False:', True and False)
print('True or False:', True or False)
print('not True:', not True)
print('2**3 =', 2**3)
print('9 // 4 =', 9 // 4)
print('9 % 4 =', 9 % 4)
````
